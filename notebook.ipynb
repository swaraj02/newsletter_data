```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
df = pd.read_csv("1stChequeNewsletter.csv")
df.info()
```

```python
plt.rcParams["figure.figsize"] = (20, 10)
```

```python
Date = df['Date created']
Campaign_Name = df['Name']
Email_Sent = df['Sent count']
Email_Read = df['Read count']
Read_Ratio = (Email_Read/Email_Sent)*100
Clicks = df['Clicks']
```

```python
fig, ax = plt.subplots(2,1, sharex=True)
plt.style.use("seaborn-bright")


ax[0].plot(df['Name'], df['Sent count'], label = "Sent Email") 
ax[0].plot(Email_Read, label = "Open Email")
ax[1].set_xlabel("Date")
ax[0].set_ylabel("Number of Emails")
ax[0].legend()
ax[0].set_title("1stCheque Newsletter")

ax[1].plot(Read_Ratio,label = "Read Ratio", color='r' )
ax[1].set_ylabel("Read Ratio (%)")
ax[1].tick_params('y')
ax[1].tick_params( 'x', labelrotation= 90)
ax[1].legend()

plt.show()
```

```python
Total_Email_Sent = df['Sent count'].sum()
Total_Email_Read = df['Read count'].sum()
Total_clicks = df['Clicks'].sum()
Total_Unsubscribed = df['Unsubscribed'].sum()
print(Total_Email_Sent)
print(Total_Email_Read)
```

```python
plt.pie([Total_Email_Sent, Total_Email_Read, Total_Unsubscribed], labels= [" Email Sent", "Email Open", "Unsubscribed"], autopct = '%.2f')
plt.legend()
plt.title("1stCheque Newsletter")
```
